```python
import numpy as np
import random
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import os
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

data_path = os.path.join(os.getcwd(), "../Traing Data Preprocessing/Training Data/Processed")

X = []
Y = []
X_val = []
Y_val = []

namesToClass = {
    "lowera": 0,
    "lowerb": 1,
    "lowerc": 2,
    "lowerd": 3,
    "lowere": 4,
    "lowerf": 5,
    "lowerg": 6,
    "lowerh": 7,
    "loweri": 8,
    "lowerj": 9,
    "lowerk": 10,
    "lowerl": 11,
    "lowerm": 12,
    "lowern": 13,
    "lowero": 14,
    "upperA": 15,
    "upperB": 16,
    "upperC": 17,
    "upperD": 18,
    "upperE": 19,
    "upperF": 20,
    "upperG": 21,
    "upperH": 22,
    "upperI": 23,
    "upperJ": 24,
    "upperK": 25,
    "upperL": 26,
    "upperM": 27,
    "upperN": 28,
    "upperO": 29,
}

for root, dirs, files in os.walk(data_path, topdown=False):
    for f in files:
        key = f[0 : 6]
        print(key)
        if key in namesToClass:
            img = load_img(os.path.join(root, f))
            arr = img_to_array(img)
            #arr.reshape((1,) + arr.shape)
            
            if random.randint(1,100) > 80:
                X_val.append(arr)
                Y_val.append(namesToClass[key])
            else:
                X.append(arr)
                Y.append(namesToClass[key])

X = np.array(X)
Y = np.array(Y)
X_val = np.array(X_val)
Y_val = np.array(Y_val)

print(X.shape)
print(Y.shape)

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(18, 13, 3)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64, activation='relu'))

model.add(Dense(30, activation='sigmoid'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=500, validation_data=(X_val, Y_val))
```
